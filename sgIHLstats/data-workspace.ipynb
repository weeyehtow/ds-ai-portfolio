```python
import numpy as np
import pandas as pd

from datetime import datetime, timedelta

print("Pandas version:", pd.__version__)
print("Numpy version:", np.__version__)
```

```python
ite = pd.read_csv("/media/user/TRANSCEND/1_Projects/SCTP-DS3F/ds-ai-portfolio/sgIHLstats/data/IntakeEnrolmentandGraduatesofITEbyCourse.csv")
poly = pd.read_csv("/media/user/TRANSCEND/1_Projects/SCTP-DS3F/ds-ai-portfolio/sgIHLstats/data/IntakeEnrolmentandGraduatesofPolytechnicsbyCourse.csv")
uni = pd.read_csv("/media/user/TRANSCEND/1_Projects/SCTP-DS3F/ds-ai-portfolio/sgIHLstats/data/IntakeEnrolmentandGraduatesofUniversitiesbyCourse.csv")

ite['IHL_type'] = 'ITE'
poly['IHL_type'] = 'Polytechnic'
uni['IHL_type'] = 'University'

combined = pd.concat([ite, poly, uni], ignore_index=True)

#print(combined)
print(combined['IHL_type'].value_counts())

#combined.describe()

df= pd.DataFrame(combined)
df

merge_keys = ['year', 'course', 'IHL_type']
value_cols = ['intake', 'enrolment', 'graduates']

mf = df[df['sex'] == 'MF'].set_index(merge_keys)[value_cols]
f  = df[df['sex'] == 'F'].set_index(merge_keys)[value_cols]

m = (mf - f).dropna().reset_index()
m['sex'] = 'M'

df = pd.concat([df, m], ignore_index=True)
df = df.rename(columns={'sex': 'gender'})
df['gender'] = df['gender'].map({'M': 'Male', 'F': 'Female', 'MF': 'Overall'})

df

#export output as csv
df.to_csv("combined.csv", header=True)
```
